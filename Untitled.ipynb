```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ls = pd.read_csv('logistics/log.log', names=['useless', 'label', 'logloss'])
ls['model'] = 'CV'


ls_bl = pd.read_csv('logistics-baseline/log.log', names=['useless', 'label', 'logloss'])
ls_bl['model'] = 'baseline'


df = pd.concat([ls, ls_bl])

df['logloss'] = df['logloss'].apply(lambda x: float(x) if x != ' False' else None)
```

```python
plt.figure(figsize=(8,15))
sns.barplot(x='logloss', y='label', data=df, hue = 'model')

plt.gcf().savefig("/mnt/z/sdfs.jpg")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("train_targets_scored.csv")
all_labels = pd.DataFrame(index = df.columns[1:].to_numpy(), )
all_labels.index.names = ['label']
```

```python
ls = pd.read_csv('logistics_new/log.log', names=['useless', 'label', 'logloss'])
ls = ls[((ls['logloss'] != ' False') & (ls['logloss'] != ' errored'))]
ls['label'] = ls['label'].str.strip(" ")
ls = ls.set_index('label')
ls = ls.drop(['useless'], axis = 1)
ls['model'] = 'tuned'
ls['logloss'] = ls['logloss'].astype(float)
```

```python
ls['logloss'].mean()
```

```python
ls_base = pd.read_csv('logistics-baseline-new/log.log', names=['useless', 'label', 'logloss'])
ls_base = ls_base[((ls_base['logloss'] != ' False') & (ls_base['logloss'] != ' errored'))]
ls_base['label'] = ls_base['label'].str.strip(" ")
ls_base = ls_base.set_index('label')
ls_base = ls_base.drop(['useless'], axis = 1)
ls_base['model'] = 'untuned'
ls_base['logloss'] = ls_base['logloss'].astype(float)
```

```python
ls_base['logloss'].mean()
```

```python
final_tuned =pd.concat([all_labels, ls], axis = 1, )
final_tuned['label index'] = range(206)

final_base = pd.concat([all_labels, ls_base], axis = 1, )
final_base['label index'] = range(206) 
```

```python
finalfinal = pd.concat([final_tuned, final_base])

finalfinal['label'] = finalfinal.index
```

```python
finalfinal
```

```python
plt.figure(figsize=(21, 11))

g = sns.barplot(x='label', y='logloss', data=finalfinal, hue = 'model',  order =all_labels.index)
# g.set(xticks=[], xlabel = '')
plt.xticks(rotation=90)
plt.tight_layout()

g.figure.savefig("/mnt/z/sdfs2dfsd.pdf")
```

```python
plt.figure(figsize=(11,7))

g = sns.barplot(x='label', y='logloss', data=finalfinal, hue = 'model', order =all_labels.index[:50] )
# g.set(xticks=[], xlabel = '')
plt.xticks(rotation=90)
plt.tight_layout()

g.figure.savefig("/mnt/z/sdfsdddd2dfsd.pdf")
```
